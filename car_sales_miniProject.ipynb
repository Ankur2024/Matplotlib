```python
%matplotlib inline 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import re
```

```python
car_info = pd.read_csv("car-sales.csv")
car_info
```

```python
 # Assuming car_info is your Pandas DataFrame
car_info["Price"] = car_info["Price"].astype(str)
car_info["Price"] = car_info["Price"].apply(lambda x: re.sub(r'\$|\.|\,', '', x))
```

```python
car_info
```

```python
#Remove Last two zeroes
car_info["Price"] = car_info["Price"].str[:-2]
car_info
```

```python
car_info["Sales Date"] = pd.date_range("1/1/2024",periods=len(car_info))
car_info
```

```python
car_info["Total Sales"] = car_info["Price"].astype(int).cumsum()
car_info
```

```python
#Let's Plot the total Sales
car_info.plot(x="Sales Date",y="Total Sales");
```

```python
car_info["Price"] = car_info["Price"].astype(int)
car_info.plot(x = "Odometer (KM)",y="Price",kind = "scatter");
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease.plot(y="age",kind="hist");
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"]>=50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind = "scatter",x = "age" , y = "chol", c = "target");
```

```python
# OO Method
fix,ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = "scatter",x = "age" , y = "chol", c = "target",ax =ax);
# ax.set_xlim([45,100])
```

```python
fix,ax=plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

# Customize the plot
ax.set(title="Heart Disease and cholestrol Level",
      xlabel = "Age",
      ylabel = "Cholestral");
```

```python
# Create subplots
fig, ax = plt.subplots(figsize=(10, 6))

# Scatter plot
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Level",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add horizontal Line
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
# Create subplots
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol levels",
        xlabel="Age",
        ylabel="Cholesterol")

# Add legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a mean line for cholesterol
ax0.axhline(y=over_50["chol"].mean(), linestyle='--', color='red', label='Mean Cholesterol')

# Add data to ax1
scatter_2 = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Thalach levels",
        xlabel="Age",
        ylabel="Thalach")

# Add legend
ax1.legend(*scatter_2.legend_elements(), title="Target")

# Add a mean line for thalach
ax1.axhline(y=over_50["thalach"].mean(), linestyle='--', color='black', label='Mean Thalach');

# Add a tile to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
