```python
prices = {
    "Mustang": 10000,
    "Benz": 20000,
    "Bmw": 5000,
    "Audi": 1000,
    "Porche": 12000
}

fig, ax = plt.subplots()

# Extract keys and values from the dictionary
cars = list(prices.keys())
prices_values = list(prices.values())

# Plot the scatter plot
ax.scatter(x=cars, y=prices_values)
# Labeling the plot
ax.set(title="Car Prices",
       xlabel="Car Model",
       ylabel="Price ($)")
plt.xticks(rotation=45)  
```

```python
fix,ax = plt.subplots()
ax.bar(cars,prices_values)
```

```python
fix,ax = plt.subplots()
ax.barh(cars,prices_values)
```

```python
#Histograph
x = np.random.randn(1000)
```

```python
fig,ax = plt.subplots()
ax.hist(x)
```

```python
#Two option for subplots
# option:1
fix,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x/2)
ax2.plot(x,x*2)
ax3.bar(cars,prices_values)
ax4.hist(x)
```

```python
#Option 2:
fix,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#plot to each different index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.randn(10),np.random.randn(10))
ax[1,0].bar(cars,prices_values)
ax[1,1].hist(x)
```
